```python
import  numpy as np
import pandas as pd
```

```python
df = pd.read_csv('us_cities_states_counties.csv',sep='delimiter', header=None)
df
```

```python
#kk1=str(kk).split('|')
cityName = []
for n in range(0,63209):
    kk=df.iloc[n]
    kk1=str(kk).split('|')
    result = ''.join([i for i in kk1[0] if not i.isdigit()])
    result=result.strip()
    cityName.append(result)
cityName
```

```python
suoxieState=[]
for n in range(0,63209):
    kk=df.iloc[n]
    kk1=str(kk).split('|')
    suoxieState.append(kk1[1])
suoxieState
```

```python
fullState=[]
for n in range(0,63209):
    kk=df.iloc[n]
    kk1=str(kk).split('|')
    fullState.append(kk1[2])
fullState
```

```python
first = pd.DataFrame({'city': cityName,
                      'abbre_state': suoxieState,
                      'states': fullState    
})
first
```

```python
second = first
second = second.drop_duplicates()
second
```

```python
second.to_csv('abbstate_city_state.csv')
```
